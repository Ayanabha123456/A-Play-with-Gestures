```python
import sklearn
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import tensorflow as tf
import numpy as np
from keras.utils import np_utils
import matplotlib.pyplot as plt
```

### train(), test() - uses data provided to class during initialization
### train_seperate(), test_seperate() - uses data directly provided to these functions during call

```python
class ANN:
    def __init__(self,data,test_size):
        self.ts=test_size
        self.df=data
        self.le=LabelEncoder()
        self.checkpoint_filepath='Weights\\best.h5'
        #building the model
        init=tf.keras.initializers.he_uniform(seed=5)
        #ANN model - ISL digits
        '''
        model = tf.keras.models.Sequential()
        model.add(tf.keras.layers.Dense(units=6,activation='relu'))
        model.add(tf.keras.layers.Dense(units=6,activation='relu'))
        model.add(tf.keras.layers.Dense(units=len(np.unique(data.iloc[:,-1])), activation='softmax'))
        '''
        #ANN model - Cambridge dataset - Hand coordinates
        '''
        model = tf.keras.models.Sequential()
        model.add(tf.keras.layers.Dense(units=32,kernel_initializer=init,activation='relu'))
        model.add(tf.keras.layers.Dense(units=32,kernel_initializer=init,activation='relu'))
        model.add(tf.keras.layers.Dense(units=len(np.unique(data.iloc[:,-1])), activation='softmax'))
        '''
        #ANN model - Cambridge dataset - Convolutional features + PCA
        '''
        model = tf.keras.models.Sequential()
        model.add(tf.keras.layers.Dense(units=32,kernel_initializer=init,activation='relu'))
        model.add(tf.keras.layers.Dense(units=32,kernel_initializer=init,activation='relu'))
        model.add(tf.keras.layers.Dense(units=32,kernel_initializer=init,activation='relu'))
        model.add(tf.keras.layers.Dense(units=16,kernel_initializer=init,activation='relu'))
        model.add(tf.keras.layers.Dense(units=len(np.unique(data.iloc[:,-1])), activation='softmax'))
        '''
        #ANN model - Cambridge dataset - Convolutional features + finger angles + PCA
        '''
        model = tf.keras.models.Sequential()
        model.add(tf.keras.layers.Dense(units=32,kernel_initializer=init,activation='relu'))
        model.add(tf.keras.layers.Dense(units=32,kernel_initializer=init,activation='relu'))
        model.add(tf.keras.layers.Dense(units=16,kernel_initializer=init,activation='relu'))
        model.add(tf.keras.layers.Dense(units=16,kernel_initializer=init,activation='relu'))
        model.add(tf.keras.layers.Dense(units=len(np.unique(data.iloc[:,-1])), activation='softmax'))
        '''
        #ANN model - Cambridge dataset - CNN features on hand edges + PCA, L.C. of BRISK, CNN features on BRISK + PCA;
        #ASL alphabet - Hand coordinates, Convolutional features + PCA
        '''
        model = tf.keras.models.Sequential()
        model.add(tf.keras.layers.Dense(units=64,kernel_initializer=init,activation='relu'))
        model.add(tf.keras.layers.Dense(units=64,kernel_initializer=init,activation='relu'))
        model.add(tf.keras.layers.Dense(units=64,kernel_initializer=init,activation='relu'))
        model.add(tf.keras.layers.Dense(units=len(np.unique(data.iloc[:,-1])), activation='softmax'))
        '''
        #ANN model- ASL alphabet - Convolutional features + finger angles + PCA, CNN features on hand edges + PCA, L.C. of BRISK, 
        #CNN features on BRISK + PCA;
        #ASL fingerspell dataset
        '''
        model = tf.keras.models.Sequential()
        model.add(tf.keras.layers.Dense(units=64,kernel_initializer=init,activation='relu'))
        model.add(tf.keras.layers.Dense(units=64,kernel_initializer=init,activation='relu'))
        model.add(tf.keras.layers.Dense(units=64,kernel_initializer=init,activation='relu'))
        model.add(tf.keras.layers.Dense(units=64,kernel_initializer=init,activation='relu'))
        model.add(tf.keras.layers.Dense(units=len(np.unique(data.iloc[:,-1])), activation='softmax'))
        '''
        #compiling the model
        model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
        self.model=model
        
    def train(self):
        #train-test split
        X_train, X_test, y_train, y_test = train_test_split(self.df.iloc[:,1:-1], self.df.iloc[:,-1], test_size=self.ts, random_state=42)
        #encoding labels
        y_train = self.le.fit_transform(y_train)
        y_train = np_utils.to_categorical(y_train)
        
        self.test_X = X_test
        self.test_y = y_test
        
        y_test = self.le.fit_transform(y_test)
        y_test = np_utils.to_categorical(y_test)
        #training the model
        callbacks = [tf.keras.callbacks.ModelCheckpoint(filepath=self.checkpoint_filepath,save_weights_only=True,monitor='val_accuracy',mode='max',save_best_only=True)]
        self.history=self.model.fit(X_train,y_train,validation_data=(X_test,y_test),batch_size = 16, epochs = 300,callbacks=callbacks)
        
    def test(self):
        #encoding labels for test data
        y=self.le.transform(self.test_y)
        y=np_utils.to_categorical(y)
        #predicting on test data
        self.model.load_weights(self.checkpoint_filepath)
        results=self.model.evaluate(self.test_X,y)
        print(results)
    
    def train_seperate(self,X_train,y_train,X_test,y_test):
        #encoding labels
        y_train = self.le.fit_transform(y_train)
        y_train = np_utils.to_categorical(y_train)
        
        self.test_X = X_test
        self.test_y = y_test
        
        y_test = self.le.fit_transform(y_test)
        y_test = np_utils.to_categorical(y_test)
        #training the model
        callbacks = [tf.keras.callbacks.ModelCheckpoint(filepath=self.checkpoint_filepath,save_weights_only=True,monitor='val_accuracy',mode='max',save_best_only=True)]
        self.history=self.model.fit(X_train,y_train,validation_data=(X_test,y_test),batch_size = 16, epochs = 100,callbacks=callbacks)
    
    def test_seperate(self,X_test,y_test):
        #encoding labels for test data
        y_test=self.le.transform(y_test)
        y_test=np_utils.to_categorical(y_test)
        #predicting on test data
        results=self.model.evaluate(X_test,y_test)
        print(results)
    
    def disp_accuracy(self):
        plt.plot(self.history.history['accuracy'])
        plt.plot(self.history.history['val_accuracy'])
        plt.title('CNN features on BRISK image + PCA-ASL fingerspell dataset-ANN-accuracy')
        plt.ylabel('accuracy')
        plt.xlabel('epoch')
        plt.legend(['train', 'test'], loc='upper left')
        plt.savefig('D://Capstone utilities//Graphs//Accuracies//CNNBRISK_ASL_fingerspell_ANN.png')
        plt.clf()
        
    def disp_loss(self):
        plt.plot(self.history.history['loss'])
        plt.plot(self.history.history['val_loss'])
        plt.title('CNN features on BRISK image + PCA-ASL fingerspell dataset-ANN-loss')
        plt.ylabel('loss')
        plt.xlabel('epoch')
        plt.legend(['train', 'test'], loc='upper left')
        plt.savefig('D://Capstone utilities//Graphs//Losses//CNNBRISK_ASL_fingerspell_ANN.png')
        plt.clf()
```
